```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Define the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the 'Company' column from 'gic' DataFrame
tfidf_matrix = tfidf_vectorizer.fit_transform(gic['Company'])

# Function to find the best match using TF-IDF
def find_best_match(asset_name):
    # Transform the asset name into a TF-IDF vector
    asset_vector = tfidf_vectorizer.transform([asset_name])
    
    # Calculate cosine similarities between the asset vector and all company vectors
    similarities = cosine_similarity(asset_vector, tfidf_matrix)
    
    # Find the index of the most similar company
    best_match_index = similarities.argmax()
    
    # Get the corresponding company details
    best_match_company = gic.loc[best_match_index, 'Company']
    best_match_gics_code = gic.loc[best_match_index, 'GICSCode']
    best_match_gic_subindustry = gic.loc[best_match_index, 'GicSubIndustry']
    
    # Calculate the accuracy (cosine similarity score)
    accuracy = similarities[0, best_match_index]
    
    return best_match_company, best_match_gics_code, best_match_gic_subindustry, accuracy

# Apply the find_best_match function to each row in 'Asset_Name' column
assets[['BestMatch', 'GICSCode', 'GicSubIndustry', 'Accuracy']] = assets['Asset Name'].apply(lambda x: pd.Series(find_best_match(x)))
```
